```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) 
```

 #Distance travelled by robber: d = 4t

```python
d_r = 4 * t
```

#Distance travelled by sheriff: d = 10(t-10)

```python
d_s = 10 * (t-10)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time(in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 20])
ax.set_ylim([0, 90])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='red')
plt.axvline(x=16.7, color = 'pink', linestyle='--')
_ = plt.axhline(y=67, color = 'pink', linestyle='--')
```
