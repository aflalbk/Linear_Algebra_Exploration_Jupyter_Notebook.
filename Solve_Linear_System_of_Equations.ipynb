####  Substitution method

     y = 3x
    -5x+2y = 2
 y in 2nd equation y = (2+5x)/2
                    = 1 + (5x/2)

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,1000)
```

```python
y1 = 3 * x
```

```python
y2 = 1 + ((5*x)/2)
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim(0,3)
ax.set_ylim(0,8)
ax.plot(x, y1, color = "green")
ax.plot(x, y2, color = "red")
plt.axvline(x=2, color = "yellow", linestyle = "dotted")
_ = plt.axhline(y=6, color = "yellow", linestyle = "dotted")
```

#### Elimination 

     2x - 3y = 15
     4x + 10y = 14

     3y = 2x-15
     y = 2x/3 - 5

     10y = 14 - 4x
     y = (14 - 4x )/10

```python
y_1 =  (2*x/3) - 5
```

```python
y_2 = (14 - (4*x) )/10
```

```python
fig, ax = plt.subplots()

plt.xlabel('x')
plt.ylabel('y')

ax.set_xlim([-5,10])
ax.set_ylim([-10,5])

ax.plot(x, y_1, color = "green")
ax.plot(x, y_2, color = "red")

plt.axvline(x=0, color = "black", linestyle = "-")
_ = plt.axhline(y=0, color = "black", linestyle = "-")

plt.axvline(x=6, color = "yellow", linestyle = "dotted")
_ = plt.axhline(y=-1, color = "yellow", linestyle = "dotted")
```
